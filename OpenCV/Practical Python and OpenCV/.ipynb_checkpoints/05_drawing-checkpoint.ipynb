```python
import numpy as np
import cv2
```

```python
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
green = (0,255,0)
cv2.line(canvas, (0,0), (300,300), green)
cv2.imshow('Line', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
red = (0,0,255)
cv2.line(canvas, (300,0), (0,300), red)
cv2.imshow('Line', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.rectangle(canvas, (10,10), (60,60), green)
cv2.imshow('Rectagle', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.rectangle(canvas, (50,200),(200,225), red, 5)
cv2.imshow('Rectangle', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blue = (255,0,0)
cv2.rectangle(canvas, (200,50),(255,125), blue, -1)
cv2.imshow('Blue Rectangle', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
canvas = np.zeros((300, 300, 3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (225,255,255)
for r in range(0, 175, 25):
    cv2.circle(canvas, (centerX, centerY), r, white)

cv2.imshow("Canvas", canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Generate a random circles on the Canvas

canvas = np.zeros((300, 300, 3), dtype="uint8")
for i in range(0, 25):
    radius = np.random.randint(5, high=10)
    color = np.random.randint(0, high=256, size=(3, )).tolist()
    pt = np.random.randint(0, high=300, size=(2, ))
    cv2.circle(canvas, tuple(pt), radius, color, -1)

cv2.imshow("Canvas", canvas)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
