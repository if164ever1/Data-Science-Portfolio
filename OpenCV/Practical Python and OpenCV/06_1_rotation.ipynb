```python
import numpy as np
import cv2
```

```python
def show_image(image, msg):
    cv2.imshow(msg, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
image = cv2.imread('./1.webp')
```

```python
show_image(image, "Original")
```

```python
(h,w) = image.shape[:2]
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
show_image(rotated, "Rotated by 45 Degrees")
```

```python
M = cv2.getRotationMatrix2D(center, -90, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
show_image(rotated, "Rotated by -90 Degrees")
```
