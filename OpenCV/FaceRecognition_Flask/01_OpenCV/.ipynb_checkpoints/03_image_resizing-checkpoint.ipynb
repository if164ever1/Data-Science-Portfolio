```python
import numpy as np
import cv2
```

```python
#Loading image
img = cv2.imread('./flying_bird.jpg')

print(img.shape)
cv2.imshow('original', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 1. Shink image
img_shink = cv2.resize(img, dsize=(320, 240), interpolation=cv2.INTER_AREA) #width - columns, hight - rows
print(img_shink.shape)
```

```python
# 2. Enlarge image
img_zoom = cv2.resize(img, dsize=(1280, 960), interpolation=cv2.INTER_CUBIC)
print(img_zoom.shape)
```

```python
cv2.imshow('Original', img)
cv2.imshow('Shink', img_shink)
cv2.imshow('Enlarge', img_zoom)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
